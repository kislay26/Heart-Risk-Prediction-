# Data Preparation

```python
data['sex'].replace({'M': 1, 'F': 0}, inplace=True)
data.chestpaintype.replace({'ASY': 0, 'NAP': 1, 'ATA': 2, 'TA': 3}, inplace=True)
data.restingecg.replace({'Normal': 0, 'LVH': 1, 'ST': 2}, inplace=True)
data['exerciseangina'].replace({'Y': 1, 'N': 0}, inplace=True)
data.st_slope.replace({'Flat': 0, 'Up': 1, 'Down': 2}, inplace=True)
data.head()
```

```python
plt.title('Distribution plots of all features')
for cols in data.columns:
    sns.distplot(data[cols])
    plt.show()
    
```

- All columns are asymptotically gaussian

```python
# shows how each feature is related with other features.
plt.figure(figsize=(18,15))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
X = data.drop('heartdisease', axis=1)
Y = data['heartdisease'].values
Y.shape
```

```python
dtr = DecisionTreeClassifier()
dtr.fit(X, Y)

plt.figure(figsize=(18,18))
feature_imp = pd.Series(dtr.feature_importances_, index=X.columns)
#feature_imp.nlargest(len(X.columns)).plot(kind='barh')
feature_imp.plot(kind='bar')
plt.show()
```

## Train Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=20, test_size=0.25)
```

## Model Training and validation

```python
lr = LogisticRegression(random_state=20)
sv = SVC(random_state=20, probability=True)
rf = RandomForestClassifier(random_state=20)
dt = DecisionTreeClassifier(random_state=20)
xgb = xgb(random_state=20)
gb = GradientBoostingClassifier(random_state=20)
models = [("Logistic_Regression", lr),
          ("Support vectors", sv),
          ("Random Forest", rf),
          ("Decision Trees", dt),
          ("XGBoost", xgb),
          ('Gradient Boosting', gb)
         ]
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.metrics import recall_score,roc_auc_score
```

```python
results = []
names = []
finalresult=[]

for name, model in models:
        kfold = KFold(n_splits=10)
        cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring="accuracy")
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)
```

- Random Forest has the highest accuracy of 86.48%

```python
# open a file, where you ant to store the model
file = open('heart_risk.pkl', 'wb')

# dump information to that file
pickle.dump(rf, file)
```

### Random Forest Hyperparameter Tuning

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 100, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'bootstrap': bootstrap}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_grid = RandomizedSearchCV(estimator = rf,param_distributions=grid, cv = 10, verbose=4, n_jobs = -1)
```

```python
opt_model = rf_grid.fit(X_train, Y_train)
```

```python
rf_grid.best_params_
```

- We use the best parameters to get the accuracy of the model.

```python
rf_opt = RandomForestClassifier(n_estimators=200, max_depth=32, max_features='sqrt', min_samples_leaf=4, min_samples_split=5,bootstrap=True)
rf_model=rf_opt.fit(X_train, Y_train)
```

```python
rf_opt_preds = rf_model.predict(X_test)
rf_opt_preds[:5], Y_test[:5]
```

```python
accuracy_score(Y_test, rf_opt_preds)
```

- There is no much improvement in the accuracy after the hyperparameter tuning
